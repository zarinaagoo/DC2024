```python
import pandas as pd
mypath = 'data/Информационно-коммуникационные технологии и связь/Уровень цифровой грамотности населения.xlsx'
df_file = pd.read_excel(mypath, skiprows=5)
df_file.rename(columns={df_file.columns[0]: 'Region'}, inplace=True)
print(df_file)
```

```python
import plotly.express as px

df_file.rename(columns={df_file.columns[0]: 'Region'}, inplace=True)

df_file_filtered = df_file[['Region'] + [col for col in df_file.columns if '6-74' in col]]

melted_df = df_file_filtered.melt(id_vars=['Region'], var_name='Year', value_name='Digital Literacy Level')

fig = px.bar(melted_df, x='Year', y='Digital Literacy Level', color='Region', barmode='group',
             title='Digital Literacy Level by Region for Ages 6-74 Over the Years')

fig.update_layout(
    updatemenus=[
        dict(
            buttons=[
                        dict(label="All Regions",
                             method="update",
                             args=[{"visible": [True] * len(melted_df['Region'].unique())},
                                   {"title": "Цифровая грамотность по региону для возрастов 6-74"}])
                    ] + [
                        dict(label=region,
                             method="update",
                             args=[{"visible": melted_df['Region'] == region},
                                   {"title": f"Уровень цифровой грамотности для региона {region} по годам"}])
                        for region in melted_df['Region'].unique()
                    ],
            direction="down",
            showactive=True
        )
    ],
    xaxis_title='Year',
    yaxis_title='Digital Literacy Level',
    template='plotly_white',
    height=600,
    width=1000
)

fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd


# Extract unique years from the DataFrames
years = [col for col in d['Jan-March'].columns if isinstance(col, int)]

# Create a figure
fig = go.Figure()

# Create a trace for each period and year
for year in years:
    visible = year == years[0]  # Only show the first year's data by default
    for period, df in d.items():
        df_filtered = df[df['Region'] != 'Республика Казахстан']
        fig.add_trace(go.Bar(
            x=df_filtered['Region'],
            y=df_filtered[year],
            name=f"{period} ({year})",
            visible=visible
        ))

# Create update menu for dropdown selection
updatemenus = [
    {
        "buttons": [
            {
                "label": str(year),
                "method": "update",
                "args": [
                    {"visible": [(year == selected_year) for selected_year in years for _ in d]},
                    {"title": f"Sum of Money by Region for Selected Year ({year})"}
                ]
            }
            for year in years
        ],
        "direction": "down",
        "showactive": True
    }
]

# Update layout
fig.update_layout(
    updatemenus=updatemenus,
    title=f"Sum of Money(x1000) by Region for Selected Year ({years[0]})",
    xaxis_title="Region",
    yaxis_title="Sum of Money",
    barmode="group",
    legend_title="Period",
    template="plotly"
)

# Show the plot
fig.show()
```
